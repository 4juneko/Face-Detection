```python
import cv2
```

```python
file_name = '1684891917163977861.png'
input_path = 'C:/Users/pc3/Desktop/unidentify_10000ea/NAVYA_un/NAVYA/rear/'
save_path = 'C:/Users/pc3/Desktop/unidentify_10000ea/NAVYA_un/NAVYA/rear/'
```

```python
img = cv2.imread(input_path + file_name)
```

```python
x, y, w, h = cv2.selectROI('location', img, False)
print('x, y:', x, y)
print('width, height: ', w, h)
```

```python
mosaic = img[y:y+h, x:x+w]
mosaic = cv2.blur(mosaic,(50,50))
img_w_mosaic = img
img_w_mosaic[y:y+h, x:x+w] = mosaic
cv2.destroyAllWindows()
cv2.imshow('mosaic', img_w_mosaic)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
save_path = save_path + file_name
cv2.imwrite(save_path, img_w_mosaic)
```
